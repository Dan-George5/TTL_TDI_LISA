```python
import numpy as np
from numpy import cos, sin, pi, exp
import matplotlib.pyplot as plt
```

```python
A    = 10e-9
T    = 600
w0   = 2*pi*15e-3
freq = np.logspace(-5, 0, 100)
w    = 2*pi*freq

phi_modsq = (A**2)*(((w0-w)**2)*((w0-w)**2)/((w0-w)*(w0+w)) + 2*cos(2*w0*T))/(16*(w0-w)*(w0+w))
print(phi_modsq)
```

```python
fig = plt.figure(figsize=(15, 10))
plt.loglog(freq, phi_modsq**0.5)
plt.xlim([1e-5, 1])
plt.axvline(x=w0/(2*pi), ls='--')
```

```python
t = np.linspace(-100, 100, 1000)
test = sin(w0*t)*sin(w0*t)

plt.plot(t, test)
```
